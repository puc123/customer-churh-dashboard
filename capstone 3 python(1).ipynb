```python
def hist_box(col):
    plt.figure(figsize=(10,2))
    plt.subplot(1,2,1)
    plt.title(f'{col}- Histogram')
    sns.histplot(data[col],kde=False)

    plt.subplot(1,2,2)
    plt.title(f'{col}- BoxPlot')
    sns.boxplot(x=data[col],color='purple',whis=True,fliersize=5)
    plt.show()
```

```python
numerical_col=data.select_dtypes(exclude='object').columns
```

```python
for col in numerical_col:
    hist_box(col)
```
